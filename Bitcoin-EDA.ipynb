```python
db=data.dropna()
```

```python
db.isnull().sum()
```

```python
db['year'],db['month'] = db.unix_conversion.dt.year, db.unix_conversion.dt.month

```

```python
db.year.value_counts()
```

```python
db.describe()
```

```python
db_2021=db[db.year==2021]
```

```python
plt.figure(figsize=(20, 10))
# addind first subplot
plt.subplot(131)
sns.barplot(x=db_2021.month,y=db_2021['Volume_(BTC)'])

# addding second subplot
plt.subplot(132)
sns.barplot(x=db_2021.month,y=db_2021['Volume_(Currency)'])

plt.subplot(133)
sns.barplot(x=db_2021.month,y=db_2021['Weighted_Price'])
```

```python
dp=db.groupby(["year"])['Volume_(BTC)'].sum().reset_index()
dp

```

```python
pctChange = dp['Volume_(BTC)'].pct_change(periods=-1)
pctChange
```
